```python
import random

def displayboard(board):
    print("---------------")
    for row in board:
        print("|", end=" ")
        for cell in row:
            print(cell, "|", end=" ")
        print("\n-------------")
def checkwin(board, player):
    for i in range(3):
        if all(cell == player for cell in board[i]) or all(board[j][i] == player for j in range(3)):
            return True
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True
    return False
def checkdraw(board):
    return all(cell != " " for row in board for cell in row)
def getmove():
    while True:
        try:
            row = int(input("Enter row number (0, 1, or 2): "))
            col = int(input("Enter column number (0, 1, or 2): "))
            if 0 <= row <= 2 and 0 <= col <= 2:
                return row, col
            else:
                print("Your selection is invalid. If you want to play a game row and column numbers must be between 0 and 2.")
        except ValueError:
            print("Your selection is invalid.you should select only integers ")
def minimax(board, depth, maximisingplayer, alpha, beta):
    if checkwin(board, "X"):
        return -10 + depth, None
    elif checkwin(board, "O"):
        return 10 - depth, None
    elif checkdraw(board):
        return 0, None
    
    if maximisingplayer:
        maximumeval = float('-inf')
        bestmove = None
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = "O"
                    eval, _ = minimax(board, depth + 1, False, alpha, beta)
                    board[i][j] = " "
                    if eval > maximumeval:
                        maximumeval = eval
                        bestmove = (i, j)
                    alpha = max(alpha, eval)
                    if beta <= alpha:
                        break
        return maximumeval, bestmove
    else:
        minimumeval = float('inf')
        bestmove = None
        for i in range(3):
            for j in range(3):
                if board[i][j] == " ":
                    board[i][j] = "X"
                    eval, _ = minimax(board, depth + 1, True, alpha, beta)
                    board[i][j] = " "
                    if eval < minimumeval:
                        minimumeval = eval
                        bestmove = (i, j)
                    beta = min(beta, eval)
                    if beta <= alpha:
                        break
        return minimumeval, bestmove
def aimove(board):
    _, move = minimax(board, 0, True, float('-inf'), float('inf'))
    return move
def playgame():
    board = [[" " for m in range(3)] for n in range(3)]
    displayboard(board)
    while True:
        row, col = getmove()
        if board[row][col] != " ":
            print("Sorry, that cell is already occupied. Please choose another cell")
            continue
        board[row][col] = "X"
        displayboard(board)
        if checkwin(board, "X"):
            print("Congratulations! You won the game")
            break
        if checkdraw(board):
            print(" game is draw !")
            break
        row, col = aimove(board)
        print(" move from AI ", row, col)
        board[row][col] = "O"
        displayboard(board)
        if checkwin(board, "O"):
            print("AI won the match")
            break
        if checkdraw(board):
            print("game is draw")
            break
if __name__ == "__main__":
    playgame()
```
